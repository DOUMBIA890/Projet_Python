```python
import pandas as pd

# Définir les chemins d'accès des fichiers
file1 = "C:\\Users\\DOUMBIA\\Desktop\\Base_économ\\export-2024-12-13T15_11_04.315Z.csv"
file2 = "C:\\Users\\DOUMBIA\\Desktop\\Base_économ\\export-2024-12-13T15_09_45.951Z.csv"
file3 = "C:\\Users\\DOUMBIA\\Desktop\\Base_économ\\API_SE.XPD.TOTL.GD.ZS_DS2_fr_csv_v2_45890\\API_SE.XPD.TOTL.GD.ZS_DS2_fr_csv_v2_45890.csv"
file4 = "C:\\Users\\DOUMBIA\\Desktop\\Base_économ\\API_NY.GDP.PCAP.CD_DS2_fr_csv_v2_3034\\API_NY.GDP.PCAP.CD_DS2_fr_csv_v2_3034.csv"

# Importation des fichiers CSV
data1 = pd.read_csv(file1, sep=';', skiprows=2)  # Data 1 (commence à la ligne 3)
data2 = pd.read_csv(file2, sep=';', skiprows=2)  # Data 2 (commence à la ligne 3)
data3 = pd.read_csv(file3, sep=';', skiprows=4)  # Data 3 (commence à la ligne 5)
data4 = pd.read_csv(file4, sep=';', skiprows=4)  # Data 4 (commence à la ligne 5)

# Suppression des colonnes inutiles
data1_clean = data1.loc[:, ~data1.columns.str.contains('^Unnamed')]
data2_clean = data2.loc[:, ~data2.columns.str.contains('^Unnamed')]
data3_clean = data3.loc[:, ~data3.columns.str.contains('^Unnamed')]
data4_clean = data4.loc[:, ~data4.columns.str.contains('^Unnamed')]

# Affichage des premières lignes pour vérifier l'importation
print("Data1 - Aperçu:")
print(data1_clean.head())

print("\nData2 - Aperçu:")
print(data2_clean.head())

print("\nData3 - Aperçu:")
print(data3_clean.head())

print("\nData4 - Aperçu:")
print(data4_clean.head())
```

```python
# Afficher les colonnes de data3_clean et data4_clean pour vérifier les noms
print("Colonnes de data3_clean :", data3_clean.columns)
print("Colonnes de data4_clean :", data4_clean.columns)

# Suppression des colonnes inutiles si elles existent
columns_to_drop_data3 = ['Country Code', 'Indicator Code']
columns_to_drop_data4 = ['Country Code', 'Indicator Code']

# Vérification et suppression des colonnes uniquement si elles existent
data3_clean = data3_clean.drop(columns=[col for col in columns_to_drop_data3 if col in data3_clean.columns], errors='ignore')
data4_clean = data4_clean.drop(columns=[col for col in columns_to_drop_data4 if col in data4_clean.columns], errors='ignore')

# Renommer les colonnes pour une meilleure lisibilité
data3_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)
data4_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)

# Afficher les premières lignes après nettoyage
print("\nData3 - Nettoyée:")
print(data3_clean.head())

print("\nData4 - Nettoyée:")
print(data4_clean.head())
```

```python
# Nettoyer les noms des colonnes en supprimant les guillemets et les virgules
data3_clean.columns = data3_clean.columns.str.replace('"', '').str.strip()
data4_clean.columns = data4_clean.columns.str.replace('"', '').str.strip()

# Vérification des nouvelles colonnes
print("\nColonnes de data3_clean après nettoyage:", data3_clean.columns)
print("\nColonnes de data4_clean après nettoyage:", data4_clean.columns)

# Suppression des colonnes inutiles si elles existent
columns_to_drop_data3 = ['Country Code', 'Indicator Code']
columns_to_drop_data4 = ['Country Code', 'Indicator Code']

# Suppression des colonnes
data3_clean = data3_clean.drop(columns=[col for col in columns_to_drop_data3 if col in data3_clean.columns], errors='ignore')
data4_clean = data4_clean.drop(columns=[col for col in columns_to_drop_data4 if col in data4_clean.columns], errors='ignore')

# Renommer les colonnes restantes pour plus de clarté
data3_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)
data4_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)

# Afficher les premières lignes après nettoyage
print("\nData3 - Nettoyée:")
print(data3_clean.head())

print("\nData4 - Nettoyée:")
print(data4_clean.head())
```

```python
# Nettoyer les noms des colonnes en supprimant les guillemets, les virgules et les espaces superflus
data3_clean.columns = data3_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()
data4_clean.columns = data4_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()

# Vérification des nouvelles colonnes après nettoyage
print("\nColonnes de data3_clean après nettoyage:", data3_clean.columns)
print("\nColonnes de data4_clean après nettoyage:", data4_clean.columns)

# Suppression des colonnes inutiles si elles existent
columns_to_drop_data3 = ['Country Code', 'Indicator Code']
columns_to_drop_data4 = ['Country Code', 'Indicator Code']

# Suppression des colonnes
data3_clean = data3_clean.drop(columns=[col for col in columns_to_drop_data3 if col in data3_clean.columns], errors='ignore')
data4_clean = data4_clean.drop(columns=[col for col in columns_to_drop_data4 if col in data4_clean.columns], errors='ignore')

# Renommer les colonnes restantes pour plus de clarté
data3_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)
data4_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)

# Afficher les premières lignes après nettoyage
print("\nData3 - Nettoyée:")
print(data3_clean.head())

print("\nData4 - Nettoyée:")
print(data4_clean.head())
```

```python
# Vérifier les noms des colonnes
print("\nColonnes de data3_clean après nettoyage (avant correction):")
print(data3_clean.columns)

print("\nColonnes de data4_clean après nettoyage (avant correction):")
print(data4_clean.columns)

# Séparer les noms de colonnes qui semblent fusionnés (s'il y en a)
# Si la concaténation est due à des virgules ou d'autres caractères, on peut les séparer
data3_clean.columns = data3_clean.columns.str.split(',').str[0]  # On garde la première partie
data4_clean.columns = data4_clean.columns.str.split(',').str[0]  # On garde la première partie

# Vérification après séparation des colonnes
print("\nColonnes de data3_clean après séparation:")
print(data3_clean.columns)

print("\nColonnes de data4_clean après séparation:")
print(data4_clean.columns)

# Nettoyer et renommer les colonnes de manière appropriée
data3_clean.columns = data3_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()
data4_clean.columns = data4_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()

# Supprimer les colonnes inutiles (par exemple, Country Code, Indicator Code si nécessaire)
columns_to_drop_data3 = ['Country Code', 'Indicator Code']
columns_to_drop_data4 = ['Country Code', 'Indicator Code']

data3_clean = data3_clean.drop(columns=[col for col in columns_to_drop_data3 if col in data3_clean.columns], errors='ignore')
data4_clean = data4_clean.drop(columns=[col for col in columns_to_drop_data4 if col in data4_clean.columns], errors='ignore')

# Renommer les colonnes 'Country Name' et 'Indicator Name' en 'Country' et 'Indicator' respectivement
data3_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)
data4_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)

# Afficher les premières lignes après nettoyage
print("\nData3 - Nettoyée après séparation:")
print(data3_clean.head())

print("\nData4 - Nettoyée après séparation:")
print(data4_clean.head())
```

```python
# Affichage des colonnes avant la séparation pour vérification
print("\nColonnes avant séparation (data3_clean) :")
print(data3_clean.columns)

print("\nColonnes avant séparation (data4_clean) :")
print(data4_clean.columns)

# Nous allons identifier la portion de la colonne qui correspond aux noms des variables et aux années
# Pour cela, nous allons diviser la chaîne de caractères en fonction des positions des années (par exemple 1960, 1961, etc.)

# Trouver où commence les années (c'est-à-dire les premières années de données)
years_start = [str(year) for year in range(1960, 2024)]

# Séparer les colonnes basées sur les années
def separate_columns(df, years_start):
    # Trouver la position où les années commencent
    columns = df.columns[0]
    for year in years_start:
        if str(year) in columns:
            break
    
    # Découper la colonne en utilisant ce point de séparation
    part1 = columns[:columns.index(str(year))]
    part2 = columns[columns.index(str(year)):]
    
    # Assigner les nouvelles colonnes
    new_columns = part1.split(',') + part2.split(',')
    return new_columns

# Appliquer la séparation des colonnes
data3_clean.columns = separate_columns(data3_clean, years_start)
data4_clean.columns = separate_columns(data4_clean, years_start)

# Affichage des nouvelles colonnes
print("\nColonnes après séparation (data3_clean) :")
print(data3_clean.columns)

print("\nColonnes après séparation (data4_clean) :")
print(data4_clean.columns)

# Nettoyage des colonnes
data3_clean.columns = data3_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()
data4_clean.columns = data4_clean.columns.str.replace(r'"|,', '', regex=True).str.strip()

# Renommer les colonnes 'Country Name' et 'Indicator Name' en 'Country' et 'Indicator' respectivement
data3_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)
data4_clean.rename(columns={'Country Name': 'Country', 'Indicator Name': 'Indicator'}, inplace=True)

# Afficher les premières lignes après nettoyage
print("\nData3 - Nettoyée après séparation:")
print(data3_clean.head())

print("\nData4 - Nettoyée après séparation:")
print(data4_clean.head())
```
