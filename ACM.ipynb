# KHI 2 :  pour voir les liaisons deux a deux 

```python
import scipy.stats as stats

# Calcul du test du chi carré pour chaque paire de variables catégorielles
categorical_columns = Comso_the_data.select_dtypes(include='object').columns

# Effectuer le test du chi carré pour chaque paire de variables catégorielles
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:  # Ne pas tester la variable contre elle-même
            # Créer un tableau de contingence
            contingency_table = pd.crosstab(Comso_the_data[col1], Comso_the_data[col2])
            # Effectuer le test du chi carré
            chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
            
            print(f"Test Chi carré entre {col1} et {col2}:")
            print(f"p-value = {p}")
            if p < 0.05:
                print("Il y a une relation significative entre ces variables.\n")
            else:
                print("Il n'y a pas de relation significative entre ces variables.\n")
```

# V de cramer : intensité de la relation

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

# Fonction pour calculer le V de Cramér
def cramers_v(chi2, n, min_dim):
    return np.sqrt(chi2 / (n * min_dim))

# Fonction pour calculer le T de Tschuprow
def tschuprows_t(chi2, n, k, r):
    return np.sqrt(chi2 / (n * (k - 1) * (r - 1)))

# Sélectionner les colonnes catégorielles
categorical_columns = Comso_the_data.select_dtypes(include='object').columns

# Initialiser un dictionnaire pour stocker les résultats
results = {}

# Calculer le test chi carré, V de Cramér et T de Tschuprow
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:  # Ne pas tester la variable contre elle-même
            # Créer un tableau de contingence
            contingency_table = pd.crosstab(Comso_the_data[col1], Comso_the_data[col2])
            
            # Effectuer le test du chi carré
            chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
            
            # Calculer V de Cramér
            n = contingency_table.sum().sum()  # Total des observations
            min_dim = min(len(contingency_table.columns) - 1, len(contingency_table.index) - 1)
            v_cramer = cramers_v(chi2, n, min_dim)
            
            # Calculer T de Tschuprow
            k = len(contingency_table.columns)
            r = len(contingency_table.index)
            t_tschuprow = tschuprows_t(chi2, n, k, r)
            
            # Ajouter les résultats au dictionnaire
            results[(col1, col2)] = {
                "Chi2 p-value": p,
                "V de Cramér": v_cramer,
                "T de Tschuprow": t_tschuprow
            }

# Convertir les résultats en DataFrame
results_df = pd.DataFrame.from_dict(results, orient='index')

# Afficher le tableau des résultats
print(results_df)
```

# ACM

# ✅ Étape 1 : Chargement et Préparation des Données
- Sélection des colonnes pertinentes.
- Vérification que toutes les colonnes existent dans le DataFrame.
- Conversion en chaînes de caractères car l'ACM ne fonctionne qu'avec des données catégoriques.

```python
import pandas as pd
import prince  # Librairie pour l'ACM
import matplotlib.pyplot as plt

# ✅ Liste des variables à utiliser
columns_to_use = [
    'ptt_dèj', 'goûter', 'soirée', 'après_déjeuner', 'après_dîner', 
    'tt_moment', 'maison', 'travail', 'salon_de_t', 'amis', 'resto', 'bar'
]

# ✅ Vérification des colonnes manquantes
missing_columns = [col for col in columns_to_use if col not in Comso_the_data.columns]

if missing_columns:
    print("🚨 Colonnes manquantes :", missing_columns)
else:
    # ✅ Sélection des données et conversion en chaînes de caractères
    comso_the_data_selected = Comso_the_data[columns_to_use].astype(str)
    print("✔ Données correctement chargées et prêtes pour l'ACM")
```

# ✅ Étape 2 : Application de l’ACM 
- On applique l’ACM avec 5 composantes principales pour capturer les relations entre les variables.
- L’ACM est entraînée sur les données catégoriques sélectionnées.

```python
# ✅ Initialisation de l'ACM avec 5 composantes principales
mca = prince.MCA(n_components=5, engine='sklearn', copy=True, check_input=True)

# ✅ Ajustement du modèle sur les données sélectionnées
mca = mca.fit(comso_the_data_selected)

print("✔ Analyse en Composantes Multiples (ACM) effectuée avec succès.")
```

# ✅ Étape 3 : Analyse de la Variance Expliquée
- On extrait les valeurs propres pour voir quelle proportion de l'information est expliquée par chaque composante.
- On affiche les résultats sous forme de Scree Plot pour mieux comprendre la répartition de la variance.


```python
# ✅ Récupération des valeurs propres (eigenvalues)
eigenvalues = mca.eigenvalues_

# ✅ Calcul de la proportion de variance expliquée
inertia_ratio = eigenvalues / eigenvalues.sum()

# ✅ Affichage des résultats
print("\n📊 Variance expliquée par chaque composant :")
for i, ratio in enumerate(inertia_ratio):
    print(f"🔹 Composant {i+1}: {ratio:.4f}")

# ✅ Visualisation sous forme de Scree Plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(inertia_ratio) + 1), inertia_ratio, marker='o', linestyle='--')
plt.title("Scree Plot : Variance expliquée par chaque composant")
plt.xlabel("Composant principal")
plt.ylabel("Proportion de la variance expliquée")
plt.grid(True)
plt.show()
```

# Étape 4 : Analyse des Contributions des Variables
- Extraction des contributions des variables à chaque axe.
- Affichage des variables qui influencent le plus chaque composante.
- Visualisation sous forme de graphique en barres 📊.


```python
# ✅ Récupération des contributions des variables
column_contrib = mca.column_contributions_

# ✅ Affichage des contributions des variables aux 2 premiers axes
print("\n📌 Contributions des variables aux axes principaux :")
for i in range(2):  # On affiche pour les 2 premiers axes
    print(f"\n🔹 Axe {i+1} :")
    print(column_contrib.iloc[:, i].sort_values(ascending=False))

# ✅ Visualisation des contributions sous forme de barres
plt.figure(figsize=(10, 6))
for i in range(2):  # Pour les 2 premiers axes
    plt.barh(column_contrib.index, column_contrib.iloc[:, i], label=f'Axe {i+1}')

plt.xlabel("Contribution (%)")
plt.ylabel("Variables")
plt.title("Contribution des variables aux axes principaux")
plt.legend()
plt.gca().invert_yaxis()  # Inverser l'axe pour meilleure lecture
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# ✅ Récupération des coordonnées des individus et des variables
row_coords = mca.row_coordinates(comso_the_data_selected)
col_coords = mca.column_coordinates(comso_the_data_selected)

# ✅ Projection des individus avec des points plus petits et une transparence améliorée
plt.scatter(row_coords[0], row_coords[1], 
            c='blue', alpha=0.3, s=15, label='Individus')

# ✅ Sélection de quelques individus pour afficher leur étiquette
n_samples = min(20, len(row_coords))  # Afficher au max 20 étiquettes
for i in range(n_samples):
    plt.text(row_coords.iloc[i, 0], row_coords.iloc[i, 1], 
             str(i), fontsize=10, color='blue', alpha=0.7)

# ✅ Projection des variables avec des flèches pour représenter leur influence
for i, var in enumerate(comso_the_data_selected.columns):
    plt.arrow(0, 0, col_coords.iloc[i, 0] * 1.2, col_coords.iloc[i, 1] * 1.2,
              color='red', alpha=0.8, head_width=0.02, head_length=0.02)
    plt.text(col_coords.iloc[i, 0] * 1.3, col_coords.iloc[i, 1] * 1.3, 
             var, fontsize=12, color='darkred')

# ✅ Amélioration de l'affichage
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.grid(True, linestyle=':', alpha=0.5)

plt.title("Projection combinée des individus et des variables (ACM)")
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.legend()
plt.show()
```
